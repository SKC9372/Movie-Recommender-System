```python
import re
dates = movies['id'].str.findall(r'\b\d{4}-\d{2}-\d{2}\b').sum()
```

```python
movies = movies[~movies['id'].apply(lambda x: any(date in x for date in dates))]
```

```python
movies = movies.merge(credits,on = 'id',how = 'left')
movies = movies.merge(keyword,on = 'id',how = 'left')
```

```python
def convert(obj):
    name = []
    for i in ast.literal_eval(obj):
        name.append(i['name'])
    return name
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[4].cast
```

```python
def convert_cast(obj):
    cast = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            cast.append(i['name'])
            counter += 1
        else:
            break
    return cast 
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.iloc[0].crew
```

```python
def get_director(obj):
    director = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            director.append(i['name'])
            break
    return director
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['cast'] + movies['crew'] + movies['genres'] + movies['keywords']
```
